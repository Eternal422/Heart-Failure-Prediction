```python
Features=["serum_creatinine","ejection_fraction"]
X = heart_data[Features]
y = heart_data["DEATH_EVENT"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.2, random_state = 2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

log_clf = LogisticRegression(random_state=1).fit(X_train, y_train)
log_pred = log_clf.predict(X_test)
accuracy_score(y_test, log_pred)
```

```python
f1_score(y_test, log_pred, average='weighted')
```

```python
from sklearn.metrics import confusion_matrix

conf_mat = confusion_matrix(y_test, log_pred)
print(conf_mat)
```

```python
from sklearn import svm

svm_clf = svm.SVC()
svm_clf.fit(X_train, y_train)
svm_pred = log_clf.predict(X_test)
accuracy_score(y_test, svm_pred)
```

```python
f1_score(y_test, svm_pred, average='weighted')
```

```python
conf_mat = confusion_matrix(y_test, svm_pred)
print(conf_mat)
```

```python
from sklearn.neighbors import KNeighborsClassifier
KN_clf = KNeighborsClassifier(n_neighbors=3)
KN_clf.fit(X_train, y_train)
KN_pred = log_clf.predict(X_test)
accuracy_score(y_test, KN_pred)
```

```python
f1_score(y_test, KN_pred, average='weighted')
```

```python
conf_mat = confusion_matrix(y_test, KN_pred)
print(conf_mat)
```

```python
from xgboost import XGBClassifier

xgb_clf = XGBClassifier(n_estimators=1000,booster = 'gblinear', learning_rate=0.3 )
xgb_clf.fit(X_train, y_train,early_stopping_rounds=4, 
             eval_set=[(X_test, y_test)], 
             verbose=False)
xgb_pred = xgb_clf.predict(X_test)
accuracy_score(y_test, xgb_pred)
```

```python
f1_score(y_test, xgb_pred, average='weighted')
```

```python
conf_mat = confusion_matrix(y_test, xgb_pred)
print(conf_mat)
```

```python
import lightgbm as lgb

lgb_clf = lgb.LGBMClassifier(boosting_type='gbdt',num_leaves=31, max_depth= 2 ,learning_rate =0.001, n_estimators=900 )
lgb_clf.fit(X_train, y_train)
lgb_pred=lgb_clf.predict(X_test)
accuracy_score(y_test, lgb_pred)
```

```python
f1_score(y_test, lgb_pred, average='weighted')
```

```python
conf_mat = confusion_matrix(y_test, lgb_pred)
print(conf_mat)
```
